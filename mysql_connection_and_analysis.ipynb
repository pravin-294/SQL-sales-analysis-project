```python
from sqlalchemy import create_engine
import pandas as pd

# format of address : mysql+pymysql://username:password@host:port/database_name
# MYSQL CONNECTION DETAILS
database_url ="mysql+pymysql://root:1234@localhost:3306/mysql_practice"

# CREATE ENGINE
engine = create_engine (database_url)

print ("connected to mysql database successfully")


# READ SALES DATA
query = "SELECT * FROM sales_s"
df = pd.read_sql(query, engine)

print(df.head())
```

```python
df.groupby("product_category")["price"].sum()
```

```python
df.groupby("product_category")["quantity"].sum()
```

```python
df["revenue"] = df["price"] * df["quantity"]

df.groupby("product_category")["revenue"].sum()
```
